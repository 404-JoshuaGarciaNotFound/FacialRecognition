```python
import numpy as np
import os
import cv2
from matplotlib import pyplot as plt

```

```python
path = 'SmallerDataSet(cleaned)'
pathsList = []
if not os.path.exists('CelebMeans'):
    os.mkdir('CelebMeans')
for celeb in os.listdir(path):
    image = np.zeros((250,250))
    maxPics = np.round(len(os.listdir(path+'/'+celeb)) * 0.8)
    count = 0
    for file in os.listdir(path+'/'+celeb):
        if (count >= maxPics):
            break
        print(file)
        image += cv2.imread(path+'/'+celeb+'/'+file,0)
        pathsList.append(str(path+'/'+celeb+'/'+file))
        count += 1
    image //= count
    cv2.imwrite('CelebMeans/'+celeb+'_Mean.jpg', image)
    
#plt.imshow(image, 'gray')
```

```python
def findClosestImage(inputPath):
    currErr = (250*250)**2
    currPick = ""
    newImage = cv2.imread(inputPath,0).astype(int)
    for celebMean in os.listdir('CelebMeans'):
        meanImage = cv2.imread('CelebMeans/'+celebMean,0).astype(int)
        newErr = np.linalg.norm(newImage-meanImage)
        if (newErr < currErr):
            currPick = celebMean
            currErr = newErr
    return currPick
```

```python
totalImages = 0
totalCorrect = 0

for celeb in os.listdir(path):
    for file in os.listdir(path+'/'+celeb):
        if str(path+'/'+celeb+'/'+file) not in pathsList:
            guess = findClosestImage(path+'/'+celeb+'/'+file)
            print(str(file[:-9] + ', ' + guess))
            if file[:-9] in guess:
                totalCorrect += 1
            totalImages += 1

print(totalCorrect / totalImages)
```
